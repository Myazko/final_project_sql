# 1.1 Songs Database

Take the songs dataset and turn it into a single `sqlite` database. 

```python
conn = sqlite3.connect('../data/song_data.db') # create a connection
c = conn.cursor() # create a cursor (a mechanism that enables traversal over the records in a database)

sql="""
CREATE TABLE SongTable(
	id integer PRIMARY KEY,
	song_name TEXT,
	song_popularity INTEGER,
	song_duration_ms INTEGER,
	acousticness FLOAT,
	danceability FLOAT,
	energy FLOAT,
	instrumentalness FLOAT,
	key INTEGER,
	liveness FLOAT,
	loudness FLOAT,
	audio_mode INTEGER,
	speechiness FLOAT,
	tempo FLOAT,
	time_signature INTEGER,
	audio_valence FLOAT
);
"""

c.execute(sql)

print("SongsTable has been created.")

conn.commit()

```

```python
with open ('../data/song_data.csv', 'r') as f:
    reader = csv.reader(f)
    columns = next(reader) 
    query = 'insert into SongTable({0}) values ({1})'
    query = query.format(','.join(columns), ','.join('?' * len(columns)))
    for data in reader:
        c.execute(query, data)
    conn.commit()



```

```python
#1.2.4 Write an SQL query to report songs with popularity of 99 and above. How many songs are there?
#Select all of them simple one....

pd.read_sql_query("""SELECT song_popularity, count(*) as pop_over99
                     from (
                         Select song_popularity
                         from songtable
                         where song_popularity>=99) a
                     Group by song_popularity
                     """, conn)

```

```python
#1.2.5 Create a new column that contains song duration in the following format m:s. 
# Example, the duration of Boulevard of Broken Dreams is 4:22
#dont group by, use only tto aggregaete
#figure out how many full minutes, figure out how many seconds, concatiante full mins, extract the remanining seconds
#total ms/1000, total ms

pd.read_sql_query("""
                     SELECT song_name,
                     ((song_duration_ms/1000)/60) || ':' || (song_duration_ms/1000%60) as song_duration_min_sec
                     from songtable
                     """, conn)
```

```python
#1.2.6 Create a column "MyFavSongs" that assigns a rank to song_popularity. 
# This column should contain a value of "Ew" when the popularity score is under 25, 
# "Ok" when the popularity score is under 50, "Yay" when the popularity score is under 75 and "OMG" if ths score is 75 and above
pd.read_sql_query("""SELECT song_name,
                     case 
                        when song_popularity<=25 then "Ew"
                        when song_popularity<=50 then "OK"
                        when song_popularity<=75 then "Yay" 
                        else "OMG" 
                        END as MyFavSongs
                     from songtable
                     """, conn)

```

```python
#1.2.7 Modify "MyFavSongs" so that "OMG!!!" is assigned when its current value is "OMG" 
# and either danceability is above 0.9 or energy is above 0.9. 
# "OMG!!!" songs are under 4 minutes long. 
# Call this new column "MyFavSongs1"
pd.read_sql_query("""SELECT *,
                     ((song_duration_ms/1000)/60) as song_duration_min,
                     case MyFavSongs
                        when "OMG" then "OMG!!!"
                     end as MyFavSongs1
                     from (SELECT *,
                     case 
                        when song_popularity<=25 then "Ew"
                        when song_popularity<=50 then "OK"
                        when song_popularity<=75 then "Yay" 
                        else "OMG" 
                        END as MyFavSongs
                     from songtable
                     )
                  where MyFavSongs like 'OMG' and (danceability >0.9 or energy>0.9) and song_duration_min<4
                  
                     
                     """, conn)
```

```python
#1.2.8 List songs that didn't end up in the "OMG!!!" category because they are longer than 4 minutes. 
# Which song has the longest duration among these songs?
#I used the order function and the answer is Te Boté - Remix	

pd.read_sql_query("""SELECT *,
                     ((song_duration_ms/1000)/60) as song_duration_min,
                     case MyFavSongs
                        when "OMG" then "OMG!!!"
                     end as MyFavSongs1
                     from (SELECT *,
                     case 
                        when song_popularity<=25 then "Ew"
                        when song_popularity<=50 then "OK"
                        when song_popularity<=75 then "Yay" 
                        else "OMG" 
                        END as MyFavSongs
                     from songtable
                     )
                  where MyFavSongs like 'OMG' and (danceability >0.9 or energy>0.9) and song_duration_min>4
                  order by song_duration_ms desc

                  
                     
                     """, conn)
```

```python
#1.2.9 Display how many songs fall under each category in column "MyFavSongs1" in descending order by count.

pd.read_sql_query("""SELECT
                     case MyFavSongs
                        when "OMG" then "OMG!!!"
                     end as MyFavSongs1, Count(*) as cnt
                     from (SELECT *,
                     ((song_duration_ms/1000)/60) as song_duration_min,
                     case 
                        when song_popularity<=25 then "Ew"
                        when song_popularity<=50 then "OK"
                        when song_popularity<=75 then "Yay" 
                        else "OMG" 
                        END as MyFavSongs
                     from songtable
                     )
                  where MyFavSongs like 'OMG' and (danceability >0.9 or energy>0.9) and song_duration_min>4
                  
                     
                     """, conn)
```

```python
#1.2.10 Display the minimum, maximum and average popularity scores. 
#The output of your query should contain only one row with 3 columns.

pd.read_sql_query("""SELECT song_popularity,
                     min(song_popularity),
                     max(song_popularity),
                     avg(song_popularity)
                     from songtable
                     """, conn)




```

```python
#1.2.11 Display average popularity scores and count for songs that contain "love" in their name and those that don't. 
# Is there any difference in their popularity?
# The songs without the word "love" in their title tend to be more popular in general.
#I tried to verify this with the sum function.
pd.read_sql_query("""SELECT  
                    avg(song_popularity) as not_Lovesongs_avarage, count (*) as cnt,
                    sum(song_popularity) as sumofpopularity
                    from songtable
                    where song_name not like "%love%"
                    group by song_popularity
                    order by song_popularity desc                 
                     """, conn)

```

```python
pd.read_sql_query("""SELECT
                    avg(song_popularity) as Lovesongs_avarage, count (*) as cnt,
                    sum(song_popularity) as sumofpopularity
                    from songtable
                    where song_name like "%love%"
                    group by song_popularity
                    order by song_popularity desc  
                                     
                     """, conn)
```

```python
#1.2.12 Display top 5 songs with the longest names.
pd.read_sql_query("""SELECT distinct(song_name),  length(song_name)
                    from songtable
                    ORDER BY length(song_name) DESC
                    LIMIT 5;
                                     
                     """, conn)
```
